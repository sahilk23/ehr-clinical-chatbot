```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig, TrainingArguments, DataCollatorForSeq2Seq
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, PeftModel 
from datasets import load_dataset, DatasetDict
from trl import SFTTrainer, SFTConfig
import os
# Disable distributed training completely
os.environ["WANDB_DISABLED"] = "true"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
os.environ["MASTER_ADDR"] = "localhost"
os.environ["MASTER_PORT"] = "12355"
os.environ["RANK"] = "0"
os.environ["WORLD_SIZE"] = "1"
os.environ["LOCAL_RANK"] = "0"

# Verify single GPU
import torch
print(f"Available GPUs: {torch.cuda.device_count()}")


# --- Configuration ---
model_id = "Qwen/Qwen2.5-Coder-7B-Instruct"

# --- 💾 DEFINE YOUR MODEL SAVING PATHS HERE ---
# This is where training checkpoints and logs will be saved.
training_output_dir = "Models/qwen3_training_checkpoints_3"

# This is the final directory where your fine-tuned model adapter will be saved.
final_model_dir = "Models/qwen3-coder-ehr-text2sql-final-3"

```

```python
# Quantization config for QLoRA
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
)
# LoRA config
lora_config = LoraConfig(
    r=16, 
    lora_alpha=32,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"]
)
```

```python
# --- Load Model and Tokenizer ---
print("Loading model and tokenizer...")
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=bnb_config,
    device_map="cuda:0",  #  Force everything to GPU 0
    torch_dtype=torch.bfloat16,  # Add this for consistency

)
tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

# # --- Prepare Model for Training ---
print("Preparing model for k-bit training...")
model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, lora_config)

# # Load your fine-tuned adapter
# model = PeftModel.from_pretrained(
#     model, 
#     "Models/qwen3-coder-ehr-text2sql-final-3_emergency_checkpoint"
# )


# --- Load Dataset ---
print("Loading dataset...")
full_ds = load_dataset("json", data_files="Data/training_dataset.jsonl", split="train")
print("✅ Dataset loaded successfully!")

# Create train/validation split (e.g., 95% train / 5% validation)
print("Creating train/validation split...")
split_ds = full_ds.train_test_split(test_size=0.05, seed=42)
ds = DatasetDict({
    "train": split_ds["train"],
    "validation": split_ds["test"]
})
print("✅ Dataset split complete!")
print(ds)

# Add EOS token to training examples
def add_eos_token(example):
    example['text'] = example['text'] + tokenizer.eos_token
    return example

# Apply to your dataset
dataset = ds.map(add_eos_token)
print("✅ EOS tokens added!")

# --- Initialize Trainer ---
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset["train"],        #  Use train split
    eval_dataset=dataset["validation"],    #  Use validation split
    peft_config=lora_config,
    processing_class=tokenizer,  #  This is correct for newer TRL
    args = SFTConfig(
        output_dir=training_output_dir,
        per_device_train_batch_size=1,
        per_device_eval_batch_size=1,  # Add this for validation
        gradient_accumulation_steps=4,
        learning_rate=1e-4,
        num_train_epochs=2,
        logging_steps=10,
        save_steps=100,
        eval_steps=100,  # Add this for validation evaluation
        eval_strategy="steps",  # Add this
        fp16=True,
        optim="paged_adamw_8bit",
        lr_scheduler_type="cosine",
        warmup_ratio=0.05,
        max_seq_length=1024,
        dataset_text_field = 'text',
        gradient_checkpointing=True, 
        dataloader_pin_memory=False,  # Add this
        dataloader_num_workers=0, 
        local_rank=-1,                    # Disable distributed training
        ddp_find_unused_parameters=False, # Disable DDP
    ),
)
```

```python
# --- Start Training ---
import wandb

print("🚀 Starting fine-tuning...")
try:
    trainer.train()

    print("✅ Fine-tuning complete!")
    
    # --- Save Final Model ---
    print(f"💾 Saving fine-tuned model adapter to '{final_model_dir}'...")
    trainer.save_model(final_model_dir)
    
    # --- Save Tokenizer (Important!) ---
    print("💾 Saving tokenizer...")
    tokenizer.save_pretrained(final_model_dir)
    
    print(f"✅ All done! Your final model is saved at: {final_model_dir}")
    
except Exception as e:
    print(f"❌ Training failed with error: {e}")
    # Optionally save checkpoint even if training fails
    print("Attempting to save current checkpoint...")
    trainer.save_model(f"{final_model_dir}_emergency_checkpoint")


# 2ef07f8a40b4788161a40599202ddfeeac2690a7
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import PeftModel
import json

# ==============================================================================
# STEP 1: CONFIGURATION AND SETUP
# ==============================================================================

# --- Model Configuration ---
# Use the same base model you fine-tuned from.
base_model_id = "Qwen/Qwen2.5-Coder-7B-Instruct"

# --- 💾 Your File & Model Paths ---
# This must point to the directory where your final adapter was saved.
final_model_dir = "Models/qwen3-coder-ehr-text2sql-final-3"

# Path to your schema file is needed to build the test prompts.
schema_file_path = "Data/tables.json"

# --- Quantization Configuration ---
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# ==============================================================================
# STEP 2: LOAD THE FINE-TUNED MODEL AND TOKENIZER
# ==============================================================================

print("Loading base model and tokenizer...")
base_model = AutoModelForCausalLM.from_pretrained(
    base_model_id,
    quantization_config=bnb_config,
    device_map={"": 0}
)
tokenizer = AutoTokenizer.from_pretrained(base_model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

print(f"Loading fine-tuned adapter from: {final_model_dir}")
# Load the LoRA adapter and merge it into the base model
model = PeftModel.from_pretrained(base_model, final_model_dir)
print("✅ Specialized model loaded successfully!")
```

```python

# ==============================================================================
# STEP 3: PREPARE HELPER FUNCTIONS AND TEST DATA
# ==============================================================================

# Helper function to format the schema (reused from the training script)
def format_schema(schema_data):
    db_info = schema_data[0]
    create_statements = []
    table_names = [name.lower() for name in db_info['table_names_original']]
    column_info = {name: [] for name in table_names}
    for col in db_info['column_names_original']:
        table_index, col_name = col
        if table_index >= 0:
            column_info[table_names[table_index]].append(f'  "{col_name.lower()}" TEXT')
    for table_name, cols in column_info.items():
        create_statements.append(f"CREATE TABLE {table_name} (\n" + ",\n".join(cols) + "\n);")
    return "\n".join(create_statements)

# Load the schema string once
with open(schema_file_path, 'r') as f:
    tables_data = json.load(f)
schema_string = format_schema(tables_data)

# --- Define Your Test Questions ---
# These are sample questions taken from your original dataset.
test_samples = [
    {
        "question": "How is potassium chl 40 meq / 1000 ml d5ns delivered to the body?",
        "expected_sql": "SELECT DISTINCT prescriptions.route FROM prescriptions WHERE prescriptions.drug = 'potassium chl 40 meq / 1000 ml d5ns'"
    },
    {
        "question": "How is olanzapine (disintegrating tablet) typically consumed?",
        "expected_sql": "SELECT DISTINCT prescriptions.route FROM prescriptions WHERE prescriptions.drug = 'olanzapine (disintegrating tablet)'"
    },
    {
        "question": "How much does patient 10019777 change in weight second measured on the last hospital visit compared to the first value measured on the last hospital visit?",
        "expected_sql": "SELECT ( SELECT chartevents.valuenum FROM chartevents WHERE chartevents.stay_id IN ( SELECT icustays.stay_id FROM icustays WHERE icustays.hadm_id IN ( SELECT admissions.hadm_id FROM admissions WHERE admissions.subject_id = 10019777 AND admissions.dischtime IS NOT NULL ORDER BY admissions.admittime DESC LIMIT 1 ) ) AND chartevents.itemid IN ( SELECT d_items.itemid FROM d_items WHERE d_items.label = 'daily weight' AND d_items.linksto = 'chartevents' ) AND chartevents.charttime = ( SELECT DISTINCT chartevents.charttime FROM chartevents WHERE chartevents.stay_id IN ( SELECT icustays.stay_id FROM icustays WHERE icustays.hadm_id IN ( SELECT admissions.hadm_id FROM admissions WHERE admissions.subject_id = 10019777 AND admissions.dischtime IS NOT NULL ORDER BY admissions.admittime DESC LIMIT 1 ) ) AND chartevents.itemid IN ( SELECT d_items.itemid FROM d_items WHERE d_items.label = 'daily weight' AND d_items.linksto = 'chartevents' ) ORDER BY chartevents.charttime ASC LIMIT 1 OFFSET 1 ) ) - ( SELECT chartevents.valuenum FROM chartevents WHERE chartevents.stay_id IN ( SELECT icustays.stay_id FROM icustays WHERE icustays.hadm_id IN ( SELECT admissions.hadm_id FROM admissions WHERE admissions.subject_id = 10019777 AND admissions.dischtime IS NOT NULL ORDER BY admissions.admittime DESC LIMIT 1 ) ) AND chartevents.itemid IN ( SELECT d_items.itemid FROM d_items WHERE d_items.label = 'daily weight' AND d_items.linksto = 'chartevents' ) AND chartevents.charttime = ( SELECT DISTINCT chartevents.charttime FROM chartevents WHERE chartevents.stay_id IN ( SELECT icustays.stay_id FROM icustays WHERE icustays.hadm_id IN ( SELECT admissions.hadm_id FROM admissions WHERE admissions.subject_id = 10019777 AND admissions.dischtime IS NOT NULL ORDER BY admissions.admittime DESC LIMIT 1 ) ) AND chartevents.itemid IN ( SELECT d_items.itemid FROM d_items WHERE d_items.label = 'daily weight' AND d_items.linksto = 'chartevents' ) ORDER BY chartevents.charttime ASC LIMIT 1 ) );"
    }
        
]

# ==============================================================================
# STEP 4: RUN THE TESTS AND DISPLAY RESULTS
# ==============================================================================

print("\n🚀 Starting model testing...")
print("-" * 50)

for i, sample in enumerate(test_samples):
    question = sample["question"]
    expected_sql = sample["expected_sql"]
    
    # Format the prompt exactly as it was during training
    prompt = f"""Given the database schema, write a SQL query for the following question.

### Schema:
{schema_string}

### Question:
{question}

### SQL:
"""

    # Generate the SQL query
    inputs = tokenizer(prompt, return_tensors="pt", return_attention_mask=False).to("cuda")
    outputs = model.generate(**inputs, max_new_tokens=150, eos_token_id=tokenizer.eos_token_id)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    # Extract just the SQL part from the full generated text
    try:
        generated_sql = generated_text.split("### SQL:")[1].strip()
    except IndexError:
        generated_sql = "Error: Could not parse SQL from model output."

    # Display the results for this test case
    print(f"--- Test Case #{i+1} ---")
    print(f"❓ Question: {question}")
    print(f"✅ Expected SQL: {expected_sql}")
    print(f"🤖 Generated SQL: {generated_sql}")
    
    # Compare the results
    if generated_sql.lower().strip() == expected_sql.lower().strip():
        print("🎯 Result: PERFECT MATCH!")
    else:
        print("❌ Result: MISMATCH")
    
    print("-" * 50)
```
